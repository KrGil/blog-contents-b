```python
import csv
import matplotlib.pyplot as plt
import numpy as np

file=open('population.csv', encoding='cp949')
age = csv.reader(file)

# 남여 인구수 저장
m = [0] * 20
f = [0] * 20

name=input('인구 구조를 알고 싶은 행정구역이나 행정구역 코들르 입력하세요!!\n')
# name = '황금1동'
for row in age:
    if name in row[0]:
        for i in range(11): 
            m.append(int(row[i+23].replace(',','')))
            f.append(int(row[i+126].replace(',','')))

print('연령별 남성 인구수:', m)
print('연령별 여성 인구수:', f)
plt.rc('font', family='NanumBarunGothic')
plt.title(name + '지역의 인구구조')

plt.plot(range(len(m)), m, label='남성')
plt.plot(range(len(f)), f, label='여성')

plt.xlim([20,30])

plt.xlabel('나이')
plt.ylabel('인구수')
plt.legend() 
plt.show()
```
